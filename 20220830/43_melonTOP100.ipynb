```python
import warnings
warnings.filterwarnings(action='ignore')
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
targetSite = 'https://www.melon.com/chart/index.htm'
req = requests.get(targetSite)
# <Response [406]>: 허용되지 않음, 요청한 페이지가 요청한 콘텐츠 특성으로 응답할 수 없다.
print(req)
```

```python
targetSite = 'https://www.melon.com/chart/index.htm'
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
req = requests.get(targetSite, headers=header)
print(req)
```

```python
html = req.text
# print(html)
soup = BeautifulSoup(html, 'html.parser')
# print(soup)
```

```python
# 노래 제목 클로링
titles = soup.findAll('div', {'class': 'ellipsis rank01'})
# print(titles)
# print(len(titles))

for i in range(len(titles)):
    # print('{:3d}위 {}'.format(i + 1, titles[i].text.strip()))
    # print('{:3d}위 {}'.format(i + 1, titles[i].text.split('\n')))
    print('{:3d}위 {}'.format(i + 1, titles[i].text.strip().split('\n')[0]))
```

```python
# 아티스트 크롤링
artists = soup.findAll('span', {'class': 'checkEllipsis'})
# print(artists)
# print(len(artists))

for i in range(len(artists)):
    # print('{:3d}위 {}'.format(i + 1, artists[i].text.strip()))
    # print('{:3d}위 {}'.format(i + 1, artists[i].text.split('\n')[0]))
    print('{:3d}위 {}'.format(i + 1, artists[i].text.strip().split('\n')[0]))
```

```python
print(dt.now().strftime('%Y-%m-%d %H:%M:%S'))
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    print('{:3d}위 {} - {}'.format(i + 1, artist, title))
```

```python
# 크롤링한 결과를 텍스트 파일에 저장한다.
file = open('./output/melonTOP100.txt', 'w')
file.write(dt.now().strftime('%Y-%m-%d %H:%M:%S') + '\n')
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    file.write('{:3d}위 {} - {}\n'.format(i + 1, artist, title))
file.close()
print('파일로 저장완료')
```
