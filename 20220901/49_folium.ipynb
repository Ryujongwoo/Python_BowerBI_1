```python
# folium 라이브러리를 사용하면 지도를 출력하고 지도위에 marker를 표시할 수 있다.
# folium 라이브러리를 설치하고 import 한다.
# !pip install folium
import folium
```

```python
# folium 라이브러리의 Map() 메소드의 location 속성으로 지정한 위치를 중심으로 하는 zoom_start 옵션으로 지정한 배율을 
# 사용하는 지도를 만든다.
foliumMap = folium.Map(location=[37.5697987, 126.9837167], zoom_start=18)

# folium 라이브러리의 Marker() 메소드의 location 속성으로 지정한 위치에 popup 속성으로 지정한 문자열을 표시하는
# marker를 만든다.
# 작성한 marker는 add_to() 메소드를 사용해서 folium 라이브러리로 생성한 지도에 추가한다.
# folium.Marker(location=[37.5697669, 126.98473111], popup='더조은컴퓨터아카데미학원 종로캠퍼스').add_to(foliumMap)
# folium.Marker(location=[37.56913788, 126.98526607], popup='더조은IT아카데미학원 종각점').add_to(foliumMap)

# marker를 클릭했을 때 팝업되는 말풍선의 크기가 마음에 들지 않으면 folium의 Popup() 메소드에서 maw_width 속성으로
# 말풍선의 가로 크기를 지정할 수 있다.
p = folium.Popup('더조은컴퓨터아카데미학원 종로캠퍼스', max_width=300)
folium.Marker(location=[37.5697669, 126.98473111], popup=p).add_to(foliumMap)
p = folium.Popup('더조은IT아카데미학원 종각점', max_width=300)
folium.Marker(location=[37.56913788, 126.98526607], popup=p).add_to(foliumMap)

# jupyter notebook에서 popup으로 표시되는 한글 메시지가 깨져서 보일 수 있다.
# save() 메소드로 folium을 사용해서 작성한 지도를 html 파일로 저장할 수 있다. => html 파일에서는 popup 속성에 
# 지정한 한글이 깨져보이지 않는다.
foliumMap.save('./foliumMap.html')

foliumMap
```
