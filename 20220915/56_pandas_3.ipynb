분석하기 좋은 데이터  
분석하기 좋은 데이터는 데이터 집합을 분석하기 좋은 형태로 만들어 놓은 것을 말하고 데이터 분석 단계에서 데이터 정리는 아주 중요한 작업으로 실제 데이터 분석 작업의 70% 이상을 차지하고 있는 작업이 데이터 정리 작업이다. => 데이터 전처리  
분석하기 좋은 데이터는 데이터 분석 목적에 맞는 데이터를 모아서 표(table)를 만들어야 하고 측정한 값은 행을 변수는 열로 구성을 하며 이 조건을 만족하는 데이터를 깔끔한 데이터(Tidy Data)라고 부른다.

```python
df1 = pd.read_csv('./data/concat_1.csv')
df1
```

```python
df2 = pd.read_csv('./data/concat_2.csv')
df2
```

```python
df3 = pd.read_csv('./data/concat_3.csv')
df3
```

```python
# concat() 메소드는 연결하려는 데이터프레임을 리스트에 담아서 전달하면 연결된 데이터프레임을 반환한다.
# concat([데이터프레임1, 데이터프레임2, 데이터프레임3, ...])
# 데이터프레임을 연결할 때 기본값으로 위에서 아래 방향으로 연결되고 df1, df2, df3은 열 이름이 모두 A, B, C, D로 같기때문에
# 데이테프레임을 연결한 다음에도 열이 그대로 유지되고 인덱스도 그대로 유지된다.
# row_concat = pd.concat([df1, df2, df3])
row_concat = pd.concat([df1, df2, df3], axis=0) # 열 방향으로 연결한다.
row_concat
```

```python
# 데이터프레임에 시리즈 연결하기
new_row_series = pd.Series(['n1', 'n2', 'n3', 'n4'])
new_row_series
```

```python
# 시리즈가 axis를 생략하면 새로운 행으로 추가되지 않고 새로운 열로 추가된다. 
# NaN이라는 값이 많이 생긴다. => NaN을 누락값(결측치)이라 부른다.
# pd.concat([df1, new_row_series], axis=0)
pd.concat([df1, new_row_series])
```

```python
# 행이 1개인 데이터프레임을 생성해서 df1에 추가한다.
# 데이터프레임을 만드는 DataFrame() 메소드는 데이터프레임을 구성하려는 데이터는 반드시 전체를 리스트로 묶어 전달해야 한다.
# 데이터프레임으로 구성할 데이터를 리스트로 묶어서 전달하지 않으면 데이터 하나하나를 개별 데이터로 취급해서 행 단위의
# 데이터프레임이 생성되지 않고 열 단위의 데이터프레임이 생성된다.
new_row_df = pd.DataFrame(['n1', 'n2', 'n3', 'n4']) # 리스트로 묶는다 => [['n1', 'n2', 'n3', 'n4']] <= 요렇게
new_row_df
```

```python
pd.concat([df1, new_row_df])
```

```python
new_row_df2 = pd.DataFrame([['n1', 'n2', 'n3', 'n4']]) # 리스트로 묶는다 => [['n1', 'n2', 'n3', 'n4']] <= 요렇게
new_row_df2
```

```python
pd.concat([df1, new_row_df2])
```

```python
# 열 이름을 지정하지 않으면 판다스가 '0', '1', '2', '3'과 같은 방식으로 열 이름을 붙여주기 때문에 columns 속성을
# 사용해서 데이터프레임 생성시 열 이름을 지정해야 한다.
# columns 속성만 사용하면 데이터프레임의 열 이름을 얻어오는 기능이 실행된다.
print(list(new_row_df2.columns))
new_row_df2 = pd.DataFrame([['n1', 'n2', 'n3', 'n4']], columns=['A', 'B', 'C', 'D'])
print(list(new_row_df2.columns))
```

```python
pd.concat([df1, new_row_df2])
```

```python
new_row_df3 = pd.DataFrame([['n1', 'n2', 'n3', 'n4']])
new_row_df3
```

```python
# 데이터프레임의 columns 속성만 사용하면 데이터프레임의 열 이름만 얻어오지만 columns 속성에 리스트로 묶은
# 열 이름을 전달하면 데이터프레임의 열 이름을 변경할 수 있다.
new_row_df3.columns = ['A', 'B', 'C', 'D']
print(list(new_row_df2.columns))
```

```python
pd.concat([df1, new_row_df3])
```

```python
# concat() 메소드로 데이터프레임을 연결할 때 axis=1로 지정하면 왼쪽에서 오른쪽으로 합치기가 실행된다.
col_concat = pd.concat([df1, df2, df3], axis=1)
col_concat
```

```python
col_concat.columns
```

```python
# 같은 열 이름이 있는 데이터프레임에서 열 이름으로 데이터를 추출하면 해당 열 이름의 데이터를 모두 추출한다.
col_concat['A']
```

```python
# concat() 메소드의 ignore_index 속성의 기본값은 False이고 기존 열 이름을 유지하면서 데이터프레임을 연결한다.
col_concat = pd.concat([df1, df2, df3], axis=1, ignore_index=False)
col_concat
```

```python
# concat() 메소드의 ignore_index 속성을 True로 지정하면 기존 열 이름을 무시하고 0부터 1씩 증가하면서 열 이름을
# 다시 부여한다.
col_concat = pd.concat([df1, df2, df3], axis=1, ignore_index=True)
col_concat
```

```python
col_concat.columns = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L']
col_concat
```

```python
# 데이터프레임['열이름'] = [리스트] 형태로 실행하면 간편하게 새로운 열을 추가할 수 있다.
col_concat['new_col_list'] = ['n1', 'n2', 'n3', 'n4']
col_concat
```

```python
# concat() 메소드는 한번에 2개 이상의 데이터프레임을 연결하는 메소드이다.
# 만약 연결할 데이터프레임이 1개라면 append() 메소드를 사용해도 된다.
df2.append(new_row_df3)
```

```python
# concat(), append() 메소드를 사용해서 데이터프레임을 연결할 때 인덱스를 다시 부여하려면 ignore_index
# 옵션을 사용한다.
```

```python
df2.append(new_row_df3, ignore_index=True)
```

```python
# append() 메소드에 딕셔너리를 사용하면 데이터를 데이터프레임으로 구성하지 않아도 데이터프레임에 추가할 수 있다.
# 데이터프레임.append(추가할 데이터가 저장된 딕셔너리)
data_dict = {'A': 'n5', 'B': 'n6', 'C': 'n7', 'D': 'n8'}
print(data_dict)
# append() 메소드로 딕셔너리를 데이터프레임에 추가하려면 딕셔너리는 데이터프레임을 구성하는 인덱스가 없으므로
# ignore_index=True 옵션을 지정해서 인덱스를 다시 부여해야 에러가 발생하지 않는다.
df3.append(data_dict, ignore_index=True)
```

```python
# join 속성을 'inner'로 지정해서 데이터프레임의 공통 열만 골라 연결하면 누락값이 생기지 않는다.
# join 속성의 기본값이 'outer'로 공통열 여부와 상관없이 연결한다. => NaN이 대량 발생될 수 있다.
```

```python
print(df1.columns)
df2.columns = ['E', 'F', 'G', 'H']
print(df2.columns)
df3.columns = ['A', 'C', 'F', 'H']
print(df3.columns)
```

```python
# row_concat = pd.concat([df1, df2, df3])
row_concat = pd.concat([df1, df2, df3], join='outer')
row_concat
```

```python
# df1, df2, df3 데이터프레임은 3개 모두 공통 열이 없기 때문에 빈 데이터프레임이 출력된다.
row_concat = pd.concat([df1, df2, df3], join='inner')
row_concat
```

```python
# df1, df3 데이터프레임은 2개 모두 공통 열이 있기 때문에 공통 열을 연결한다.
pd.concat([df1, df3], join='inner')
```

```python
# df2, df3 데이터프레임은 2개 모두 공통 열이 있기 때문에 공통 열을 연결한다.
pd.concat([df2, df3], join='inner')
```

```python
print(list(df1.index))
# 데이터프레임의 인덱스를 변경하려면 데이터프레임의 index 속성에 새 인덱스를 리스트로 전달하면 된다.
df2.index = [4, 5, 6, 7]
print(list(df2.index))
df3.index = [0, 2, 5, 6]
print(list(df3.index))
```

```python
col_concat = pd.concat([df1, df2, df3], axis=1)
col_concat
```

```python
# df1, df3 데이터프레임은 2개 모두 공통 행이 있기 때문에 공통 행을 연결한다.
pd.concat([df1, df3], axis=1, join='inner')
```

```python
# df2, df3 데이터프레임은 2개 모두 공통 행이 있기 때문에 공통 행을 연결한다.
pd.concat([df2, df3], axis=1, join='inner')
```
