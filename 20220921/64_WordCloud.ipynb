```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.rcParams['font.size'] = 15
```

```python
# !pip install wordcloud
from wordcloud import WordCloud
```

```python
text = open('./data/wordData.txt').read()
print(text)
```

```python
# generate() 함수로 단어별 출현 빈도를 비율로 계산한다.
wordcloud = WordCloud().generate(text)
print(wordcloud)
wordcloud.words_
```

```python
plt.figure(figsize=[15, 10])
plt.imshow(wordcloud, interpolation='bilinear') # 워드 클라우드를 만든다.
plt.axis('off') # 그래프의 눈금을 숨긴다.
plt.show()
```

<img src="./interpolation.png" width=850 align="left"/>

```python
# WordCloud() 함수에 max_font_size 옵션을 지정하면 최고로 출현 빈도가 많은 단어의 크기를 지정할 수 있다.
wordcloud = WordCloud(max_font_size=100).generate(text)
plt.figure(figsize=[15, 10])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# WordCloud() 함수에 max_words 옵션을 지정하면 워드클라우드에 표시되는 단어의 개수를 지정할 수 있다.
wordcloud = WordCloud(max_words=100).generate(text)
plt.figure(figsize=[15, 10])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from PIL import Image # 이미지에서 마스킹 데이터를 뽑아낸다.
from wordcloud import STOPWORDS # 불용어 사전 => 워드 클라우드로 작성할 단어에서 제외할 단어
```

```python
# 워드 클라우드 모양으로 사용할 이미지를 읽어 넘파이 배열로 저장한다.
mask = np.array(Image.open('./data/alice_mask.png'))
print(mask)
```
