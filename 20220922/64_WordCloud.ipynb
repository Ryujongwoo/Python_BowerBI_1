워드 클라우드 작성 순서

```python
# 워드 클라우드로 작성할 말뭉치를 읽어온다.
text = open('./data/alice.txt').read()
print(text)
```

```python
from PIL import Image # 이미지에서 마스킹 데이터를 뽑아낸다.
from wordcloud import STOPWORDS # 불용어 사전 => 워드 클라우드로 작성할 단어에서 제외할 단어
```

```python
# 워드 클라우드 모양으로 사용할 이미지를 읽어 넘파이 배열로 저장한다.
mask = np.array(Image.open('./data/alice_mask.png'))
print(mask)
```

```python
# 불용어를 처리한다.
stopwords = set(STOPWORDS)
print(type(stopwords))
stopwords
```

```python
# 불용어를 추가한다.
stopwords.add('said')
stopwords
```

```python
# 불용어를 제거한다.
stopwords.remove('said')
stopwords
```

```python
# 워드 클라우드 설정 및 생성
# background_color 속성으로 워드 클라우드의 배경색을 지정한다.
# stopwords 속성으로 불용어를 지정한다.
# mark 속성으로 워드 클라우드 모양으로 사용할 마스크를 지정한다.
wc = WordCloud(background_color='white', max_words=2000, stopwords=stopwords, mask=mask) # 워드 클라우드 설정
wordcloud = wc.generate(text)
```

```python
# 워드 클라우드를 표시한다.
plt.figure(figsize=[15, 10])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import random
```

```python
# 워드클라우드를 그레이톤으로 표시하는 함수를 만든다. => recolor() 함수의 color_func 속성에 지정한다.
# word => 워드클라우드로 표시할 단어
# font_size => 워드클라우드로 표시할 단어의 크기
# position => 워드클라우드로 표시할 단어의 위치
# orientation => 워드클라우드로 표시할 단어의 방향
# hwargs => 워드클라우드로 표시할 단어에 적용할 글꼴
def gray_color_func(word, font_size, position, orientation, random_state=None, **hwargs):
    # print(word, font_size, position, orientation, hwargs)
    return 'hsl(0, 0%%, %d%%)' % random.randint(0, 100)
```

```python
plt.figure(figsize=[15, 10])
plt.imshow(wordcloud.recolor(color_func=gray_color_func, random_state=3), interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import nltk
from konlpy.corpus import kolaw
from konlpy.tag import Twitter
t = Twitter()
```

```python
# konlpy 패키지는 대한민국 헌법 말뭉치 kolaw와 국회 법안 말뭉치 kobill을 제공한다.
ko_con_text = kolaw.open('constitution.txt').read()
print(ko_con_text)
```

```python
# 한글 데이터 형태소 분석을 실행한다.
kokens_ko = t.nouns(ko_con_text)
kokens_ko
```

```python
# 불용어를 정리한다.
stop_words = ['제', '년', '월', '일', '조', '수', '때', '그', '이', '바', '및', '정', '관', '위', '장', '안', '거나', '다만', 
              '인', '경우', '이상', '모든', '기타']
```

```python
# 형태소 분석 결과에서 불용어를 제거한다.
# 형태소 분석 결과(kokens_ko)의 각 단어가 stop_words에 없는 단어만 다시 저장한다.
kokens_ko = [each_word for each_word in kokens_ko if each_word not in stop_words]
```

```python
# 형태소 분석 결과 단어별 출현 횟수를 계산한다.
ko = nltk.Text(kokens_ko)
ko.vocab().most_common(50)
```

```python
# 단어별 출현 횟수가 많은 순서대로 워드 클라우드로 표현할 단어를 추출해서 딕셔너리 형태로 만든다.
data = ko.vocab().most_common(500)
tmp_data = dict(data)
tmp_data
```

```python
# 딕셔너리에 저장된 데이터는 generate_from_frequencies() 함수를 사용해서 워드 클라우드로 만든다.
wordcloud = WordCloud(
    font_path='c:\\windows\\fonts\\nanumgothiccoding.ttf',
    background_color='white',
    relative_scaling=0.5
).generate_from_frequencies(tmp_data)
```

```python
plt.figure(figsize=[15, 10])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
